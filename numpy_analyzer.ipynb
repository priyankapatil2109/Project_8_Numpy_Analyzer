```python
import numpy as np

class DataAnalytics:

    # ---- Constructor ----
    def __init__(self, array=None):
        self.__array = np.array(array) if array is not None else None  # Encapsulation: private attribute

    # ---- Array Management ----
    def create_array(self, data, shape=None):
        """Create 1D, 2D, or 3D NumPy array."""
        self.__array = np.array(data)
        if shape:
            self.__array = self.__array.reshape(shape)
        return self.__array

    def get_array(self):
        return self.__array

    def set_array(self, array):
        self.__array = np.array(array)

    def index_array(self, index):
        """Access specific element by index."""
        return self.__array[index]

    def slice_array(self, row_range, col_range=None):
        """Slice rows and optionally columns."""
        if col_range:
            return self.__array[row_range[0]:row_range[1], col_range[0]:col_range[1]]
        return self.__array[row_range[0]:row_range[1]]

    def combine_arrays(self, other_array, axis=0):
        """Concatenate two arrays along given axis."""
        other = np.array(other_array)
        return np.concatenate((self.__array, other), axis=axis)

    def vstack_arrays(self, other_array):
        """Vertical stack (row-wise) using np.vstack."""
        other = np.array(other_array)
        return np.vstack((self.__array, other))

    def hstack_arrays(self, other_array):
        """Horizontal stack (column-wise) using np.hstack."""
        other = np.array(other_array)
        return np.hstack((self.__array, other))

    def split_array(self, indices):
        """Split array into sub-arrays."""
        return np.split(self.__array, indices)

    # ---- Mathematical Operations ----
    def add(self, other):
        return self.__array + np.array(other).reshape(self.__array.shape)

    def subtract(self, other):
        return self.__array - np.array(other).reshape(self.__array.shape)

    def multiply(self, other):
        return self.__array * np.array(other).reshape(self.__array.shape)

    def divide(self, other):
        """Division with zero-check."""
        other_arr = np.array(other).reshape(self.__array.shape)
        if np.any(other_arr == 0):
            raise ZeroDivisionError("Division by zero detected in the second array!")
        return self.__array / other_arr

    def dot_product(self, other):
        """Dot product for 1D arrays."""
        return np.dot(self.__array, np.array(other))

    def matrix_multiply(self, other):
        """Matrix multiplication for 2D arrays."""
        return np.matmul(self.__array, np.array(other))

    # ---- Search, Sort, Filter ----
    def search(self, value):
        """Search for value; returns indices."""
        return np.where(self.__array == value)

    def sort_array(self, ascending=True, axis=-1):
        """Sort in ascending or descending order (works for any dimension)."""
        sorted_arr = np.sort(self.__array, axis=axis)
        if not ascending:
            # Flip along the same axis used for sorting
            sorted_arr = np.flip(sorted_arr, axis=axis)
        return sorted_arr

    def filter_array(self, condition_func):
        """Filter based on user-defined condition (works on flattened array)."""
        flat = self.__array.flatten()
        return flat[condition_func(flat)]

    # ---- Aggregating Functions ----
    def compute_sum(self):      return np.sum(self.__array)
    def compute_mean(self):     return np.mean(self.__array)
    def compute_median(self):   return np.median(self.__array)
    def compute_std(self):      return np.std(self.__array)
    def compute_variance(self): return np.var(self.__array)

    # ---- Statistical Functions ----
    def min_value(self):        return np.min(self.__array)
    def max_value(self):        return np.max(self.__array)

    def percentile(self, p):
        """Compute the p-th percentile."""
        return np.percentile(self.__array, p)

    def correlation(self, other):
        """Correlation coefficients between two arrays."""
        return np.corrcoef(self.__array.flatten(), np.array(other).flatten())

    # ---- OOP: Class Method & Static Method ----
    @classmethod
    def from_range(cls, start, stop, step=1):
        """Class method: create DataAnalytics object from a range."""
        return cls(np.arange(start, stop, step))

    @staticmethod
    def display_array(array):
        """Static method: display any array."""
        print(array)


print('✅ DataAnalytics class loaded successfully!')
```

```python
# ==================== HELPER FUNCTIONS ====================

def get_int_list(prompt):
    """Accept space-separated OR comma-separated integers with validation."""
    while True:
        try:
            user_input = input(prompt).replace(',', ' ')
            return list(map(int, user_input.split()))
        except ValueError:
            print("  ⚠️  Invalid input! Please enter integers only (e.g., 1 2 3 or 1,2,3).")

def get_int(prompt):
    """Read a single integer with validation."""
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            print("  ⚠️  Invalid input! Please enter a whole number.")

def get_float(prompt):
    """Read a single float with validation."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("  ⚠️  Invalid input! Please enter a number.")


# ----------------------------------------------------------------
def create_numpy_array():
    """REQUIREMENT: Array Management - Create 1D/2D/3D, Indexing, Slicing."""
    print("\nSelect the type of array to create:")
    print("1. 1D Array")
    print("2. 2D Array")
    print("3. 3D Array")
    choice = get_int("Enter your choice: ")

    da = DataAnalytics()

    try:
        if choice == 1:
            elements = get_int_list("Enter the elements separated by space: ")
            da.create_array(elements)
            print("\nArray created successfully:")
            print(da.get_array())

        elif choice == 2:
            rows = get_int("\nEnter the number of rows: ")
            cols = get_int("Enter the number of columns: ")
            n = rows * cols
            elements = get_int_list(f"Enter {n} elements for the array separated by space: ")
            da.create_array(elements, shape=(rows, cols))
            print("\nArray created successfully:")
            print(da.get_array())

        elif choice == 3:
            depth = get_int("\nEnter depth: ")
            rows  = get_int("Enter rows: ")
            cols  = get_int("Enter columns: ")
            n = depth * rows * cols
            elements = get_int_list(f"Enter {n} elements separated by space: ")
            da.create_array(elements, shape=(depth, rows, cols))
            print("\nArray created successfully:")
            print(da.get_array())

        else:
            print("Invalid choice.")
            return None

    except ValueError as e:
        print(f"  ❌ Error creating array: {e}")
        return None

    # Indexing / Slicing sub-menu
    print("\nChoose an operation:")
    print("1. Indexing")
    print("2. Slicing")
    print("3. Go Back")
    op = get_int("Enter your choice: ")

    try:
        if op == 1:
            if choice == 1:
                idx = get_int("Enter the index: ")
                print(f"\nElement at index {idx}: {da.index_array(idx)}")
            else:
                idx = tuple(get_int_list("Enter index (row col): "))
                print(f"\nElement at {idx}: {da.index_array(idx)}")

        elif op == 2:
            row_range = list(map(int, input("Enter the row range (start:end): ").split(":")))
            if choice == 1:
                sliced = da.slice_array(row_range)
            else:
                col_range = list(map(int, input("Enter the column range (start:end): ").split(":")))
                sliced = da.slice_array(row_range, col_range)
            print("\nSliced Array:")
            print(sliced)

    except (IndexError, ValueError) as e:
        print(f"  ❌ Error: {e}")

    return da


# ----------------------------------------------------------------
def perform_math_operations(da):
    """REQUIREMENT: Mathematical Operations."""
    print("\nCurrent Array:")
    print(da.get_array())
    arr_shape = da.get_array().shape
    arr_size  = da.get_array().size

    print("\nChoose a mathematical operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    op = get_int("Enter your choice: ")

    elements2 = get_int_list(f"Enter the same-size array ({arr_size} elements) separated by space: ")

    print("\nOriginal Array:")
    print(da.get_array())
    print("\nSecond Array:")
    print(np.array(elements2).reshape(arr_shape))

    try:
        if op == 1:
            result = da.add(elements2)
            print("\nResult of Addition:")
        elif op == 2:
            result = da.subtract(elements2)
            print("\nResult of Subtraction:")
        elif op == 3:
            result = da.multiply(elements2)
            print("\nResult of Multiplication:")
        elif op == 4:
            result = da.divide(elements2)   # ZeroDivisionError raised inside class
            print("\nResult of Division:")
        else:
            print("Invalid choice.")
            return
        print(result)

    except ZeroDivisionError as e:
        print(f"  ❌ Division Error: {e}")
    except ValueError as e:
        print(f"  ❌ Value Error: {e}")


# ----------------------------------------------------------------
def combine_or_split(da):
    """REQUIREMENT: Combining (vstack/hstack) and Splitting arrays."""
    print("\nCurrent Array:")
    print(da.get_array())

    print("\nChoose an option:")
    print("1. Combine Arrays (vstack - vertical/row-wise)")
    print("2. Combine Arrays (hstack - horizontal/column-wise)")
    print("3. Split Array")
    op = get_int("Enter your choice: ")

    try:
        if op == 1:
            elements2 = get_int_list("Enter elements of another array to vstack (separated by space): ")
            result = da.vstack_arrays(elements2)
            print("\nOriginal Array:")
            print(da.get_array())
            print("\nSecond Array:")
            print(np.array(elements2))
            print("\nCombined Array (vstack - Vertical Stack):")
            print(result)

        elif op == 2:
            elements2 = get_int_list("Enter elements of another array to hstack (separated by space): ")
            result = da.hstack_arrays(elements2)
            print("\nOriginal Array:")
            print(da.get_array())
            print("\nSecond Array:")
            print(np.array(elements2))
            print("\nCombined Array (hstack - Horizontal Stack):")
            print(result)

        elif op == 3:
            indices = get_int_list("Enter split indices (space-separated): ")
            parts = da.split_array(indices)
            print("\nOriginal Array:")
            print(da.get_array())
            print("\nSplit Arrays:")
            for p in parts:
                print(p)
        else:
            print("Invalid choice.")

    except ValueError as e:
        print(f"  ❌ Error: {e}")


# ----------------------------------------------------------------
def search_sort_filter(da):
    """REQUIREMENT: Search, Sort, Filter."""
    print("\nCurrent Array:")
    print(da.get_array())

    print("\nChoose an option:")
    print("1. Search a value")
    print("2. Sort the array")
    print("3. Filter values")
    op = get_int("Enter your choice: ")

    print("\nOriginal Array:")
    print(da.get_array())

    try:
        if op == 1:
            val = get_int("Enter value to search: ")
            result = da.search(val)
            if result[0].size > 0:
                print(f"\nValue {val} found at indices: {result[0]}")
            else:
                print(f"\nValue {val} not found in array.")

        elif op == 2:
            order = input("Sort ascending? (yes/no): ").strip().lower()
            ascending = order != "no"
            sorted_arr = da.sort_array(ascending=ascending)
            print("\nSorted Array:")
            print(sorted_arr)
            direction = "Ascending" if ascending else "Descending"
            print(f"(Sorting applied {direction} row-wise.)")

        elif op == 3:
            print("Enter condition (e.g.,  >30   <20   ==5   >=10   <=50)")
            condition_str = input("Condition: ").strip()
            if len(condition_str) >= 2 and condition_str[1] == '=':
                op_char   = condition_str[:2]
                threshold = float(condition_str[2:])
            else:
                op_char   = condition_str[0]
                threshold = float(condition_str[1:])

            ops_map = {
                '>':  lambda x: x > threshold,
                '<':  lambda x: x < threshold,
                '>=': lambda x: x >= threshold,
                '<=': lambda x: x <= threshold,
                '==': lambda x: x == threshold,
            }
            if op_char in ops_map:
                filtered = da.filter_array(ops_map[op_char])
                print("\nFiltered Array:")
                print(filtered)
            else:
                print("  ❌ Unsupported operator. Use: > < >= <= ==")
        else:
            print("Invalid choice.")

    except ValueError as e:
        print(f"  ❌ Error: {e}")


# ----------------------------------------------------------------
def aggregates_and_statistics(da):
    """REQUIREMENT: Aggregating Functions + Statistical Functions."""
    print("\nCurrent Array:")
    print(da.get_array())

    print("\nChoose an aggregate/statistical operation:")
    print("1. Sum")
    print("2. Mean")
    print("3. Median")
    print("4. Standard Deviation")
    print("5. Variance")
    print("6. Minimum Value")
    print("7. Maximum Value")
    print("8. Percentile")
    print("9. Correlation with another array")
    op = get_int("Enter your choice: ")

    print("\nOriginal Array:")
    print(da.get_array())

    try:
        if   op == 1: print(f"\nSum of Array: {da.compute_sum()}")
        elif op == 2: print(f"\nMean of Array: {da.compute_mean()}")
        elif op == 3: print(f"\nMedian of Array: {da.compute_median()}")
        elif op == 4: print(f"\nStandard Deviation of Array: {da.compute_std()}")
        elif op == 5: print(f"\nVariance of Array: {da.compute_variance()}")
        elif op == 6: print(f"\nMinimum Value: {da.min_value()}")
        elif op == 7: print(f"\nMaximum Value: {da.max_value()}")
        elif op == 8:
            p = get_float("Enter percentile (0-100): ")
            if not 0 <= p <= 100:
                print("  ❌ Percentile must be between 0 and 100.")
            else:
                print(f"\n{p}th Percentile: {da.percentile(p)}")
        elif op == 9:
            size = da.get_array().size
            other = get_int_list(f"Enter {size} elements for second array: ")
            if len(other) != size:
                print(f"  ❌ Please enter exactly {size} elements.")
            else:
                corr = da.correlation(other)
                print(f"\nCorrelation Coefficients:")
                print(corr)
        else:
            print("Invalid choice.")

    except ValueError as e:
        print(f"  ❌ Error: {e}")


print('✅ All helper functions loaded successfully!')
```

```python
# ==================== MAIN MENU ====================

def main():
    print("Welcome to the NumPy Analyzer!")
    print("=" * 42)

    da = None  # Shared array object reused across all operations

    while True:
        print("\nChoose an option:")
        print("1. Create a Numpy Array")
        print("2. Perform Mathematical Operations")
        print("3. Combine or Split Arrays")
        print("4. Search, Sort, or Filter Arrays")
        print("5. Compute Aggregates and Statistics")
        print("6. Exit")

        if da is not None:
            print(f"\n[Current Array: {da.get_array()}]")

        choice = get_int("Enter your choice: ")

        if choice == 1:
            da = create_numpy_array()

        elif choice in [2, 3, 4, 5]:
            if da is None:
                print("\n⚠️  No array created yet! Please choose option 1 first.")
                continue
            if   choice == 2: perform_math_operations(da)
            elif choice == 3: combine_or_split(da)
            elif choice == 4: search_sort_filter(da)
            elif choice == 5: aggregates_and_statistics(da)

        elif choice == 6:
            print("\nThank you for using the NumPy Analyzer!")
            print("Goodbye!")
            break
        else:
            print("  ⚠️  Invalid choice. Please enter a number between 1 and 6.")

main()
```
